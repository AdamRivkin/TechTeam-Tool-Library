```python
def db_print(a):
    count=0
    for i in a:
        print(i)
        for j in a[i]:
            print(f'{count}\t{j}')
            for k in a[i][j]:
                print(f'\t\t{k}\t:{a[i][j][k]}')
            count+=1
```

```python
db_print(a)
```

```python
def save_db_as_json(a,filename):
    z=open(filename+".json","w")
    json.dump(a,z)
    z.close()
    print(f"saved as {filename+'.json'}!")
```
